#### Solution to the ODE of $R$, the generating function of completely reducible trees.

$$R'(z) - \dfrac{(1-q)z^2-2z+2}{z(1-z)}R(z) = -(1-p_\epsilon).$$

```python
z = var('z')
q = var('q')
```

##### Integrate $- \dfrac{(1-q)z^2-2z+2}{z(1-z)}$

```python
P = ((1-q)*z^2 - 2*z + 2)/(z*(1-z))
mu = integral(P, z)

latex_print(mu)
```

##### Check the limit when subtracting by $1/z^2$

```python
exp_mu = exp(mu).canonicalize_radical()
latex_print(exp_mu)
```

```python
z = var('z')
q, pe = var('q pe')
R = function('R')(z)

ode = diff(R, z) - ((1-q)*z^2 - 2*z + 2)/(z*(1-z)) * R == -(1-pe)

R = desolve(ode, R, ivar=z)
latex_print(R)

R_series = taylor(R, z, 0, 2)
```
