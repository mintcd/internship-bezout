```python
from sage.all import *
```

```python
z,u,v,Z1 = var('z u v Z1')
```

```python
# Define W as a formal function of z
W = function('W')(z)

# 2. Define the Riccati substitution
Z_sub = -1/(2*u) * diff(W, z) / W

# 3. Construct the Left-Hand Side (Z')
LHS = diff(Z_sub, z)

# 4. Construct the Right-Hand Side of the Riccati equation
RHS = 2*u * (Z_sub)**2 + (2/z + 2*v*z/(1-z)) * Z_sub + (1-u-v)*z**2/(1-z)**2 - 1

# 5. Bring everything to one side (LHS - RHS = 0)
# Multiply by -2*u*W(z) to clear the denominator and linearize
linearized_expr = (LHS - RHS) * (-2*u*W)

# 6. Expand the expression to force the non-linear (W')^2 terms to cancel
expanded_expr = linearized_expr.expand()

# 7. Extract the coefficients for W'', W', and W to verify the standard form
coeff_W_bis = expanded_expr.coefficient(diff(W, z, 2)).full_simplify()
coeff_W_prime = expanded_expr.coefficient(diff(W, z)).full_simplify()
coeff_W = expanded_expr.coefficient(W).full_simplify()

# Display the results
print("Coefficient of W''(z) :", coeff_W_bis)
print("Coefficient of W'(z)  :", coeff_W_prime)
print("Coefficient of W(z)   :", coeff_W)
```
