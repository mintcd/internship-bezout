This notebook is to sample regex tree uniformly

```python
import numpy as np
import random
```

The recurrence is

$$(n+1) t_n = (2n-1) t_{n-1} + 7(n-2) t_{n-2} + 4(2n-7) t_{n-3}, \quad t_1 = 1,t_2 = 2,t_3 = 4.$$

```python
def compute_tns(N):
    tns = [0] * (N + 1)
    if N >= 1: tns[1] = 1
    if N >= 2: tns[2] = 1
    
    for n in range(3, N + 1):
        tns[n] = ((2*n - 1)*tns[n - 1] + 7*(n - 2)*tns[n - 2]) // (n + 1)
        
    return tns

def random_split(n, tns):
    m = n - 1 
    total = (tns[n] - tns[n - 1]) // 2 
    remaining = random.randrange(total)
    
    left, right = 1, m - 1
    while left <= right:
        remaining -= tns[left] * tns[m - left]
        if remaining < 0:
            return left
            
        if left != right:
            remaining -= tns[right] * tns[m - right]
            if remaining < 0:
                return right
                
        left += 1
        right -= 1

    raise ValueError("Should never reach here")

def sample_tree(n, tns):
    if n == 1:
        return "a"
        
    total_trees = tns[n]
    r = random.randrange(total_trees)
    
    weight_star = tns[n - 1]
    
    if r < weight_star:
        return ("*", sample_tree(n - 1, tns))
        
    r -= weight_star
    weight_binary = (tns[n] - tns[n - 1]) // 2
    
    op = "+" if r < weight_binary else "."
        
    left_size = random_split(n, tns)
    right_size = (n - 1) - left_size
    
    return (op, sample_tree(left_size, tns), sample_tree(right_size, tns))
```

```python
tns = compute_tns(10)
print(sample_tree(10, tns))
```
